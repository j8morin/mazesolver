```python
grid=[]
for i in range(400):
    grid.append(i)

#print(grid)
    
def getFitness(cell):
        #FV=(CC-SC)/(FC-CC) * 100 with SC=390 and FC=9
        SC=0
        FC=400
        CC=cell
        if (FC-CC==0):    #avoid division by zero
            return 40000
        else:
            fitness = abs(((CC-SC)/(FC-CC))*100)
        return fitness

fitness=[]
for i in range (len(grid)):
    f=getFitness(grid[i])
    fitness.append(f)

#print(fitness)
t=np.arange(0,400,1)

plt.plot(t,fitness)
```
